```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Count total unique players
total_unique_players = len(purchase_data["SN"].unique())
total_unique_players

#Make new dataframe to summarize total unique players
total_player_summary = pd.DataFrame({"Total Players": [total_unique_players]})
total_player_summary
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Count total unique items
total_unique_items = len(purchase_data["Item ID"].unique())
total_unique_items

#Find the average price
average_price = purchase_data["Price"].mean()
average_price

#Find number of purchases
number_of_purchases = len(purchase_data["Purchase ID"].unique())
number_of_purchases

#Find total revenue
total_revenue = purchase_data["Price"].sum()
total_revenue

#Make new dataframe to show number of unique items, avg prices, etc. 
purchasing_analysis = pd.DataFrame({"Number of Unique Items": [total_unique_items],
                                   "Average Price": [average_price],
                                   "Number of Purchases": [number_of_purchases],
                                   "Total Revenue": [total_revenue]})                                
purchasing_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Locate specific customer purchase data to remove duplicates from
customer_purchases = purchase_data.loc[:,["Gender", "SN","Age"]]
customer_purchases

#Remove the duplicates from purchase data
purchase_data_nodups = customer_purchases.drop_duplicates()
purchase_data_nodups

#Count of players by gender without duplicates
total_gender = purchase_data_nodups["Gender"].value_counts()
total_gender

#Percentage of players by gender
gender_percentage = total_gender / total_unique_players * 100
gender_percentage

#Gender demographics DataFrame
gender_demographics = pd.DataFrame({"Total Count": total_gender,
                                   "Percentage of Players": gender_percentage.astype(float).map("{:,.2f}%".format)}
                                    )
gender_demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#purchase count by gender
gender_purchases = purchase_data["Gender"].value_counts()
gender_purchases

#avg purchase price by gender
gender_averages = purchase_data.groupby("Gender").mean()["Price"]
gender_averages

#total purchase value by gender
gender_purchase_val = purchase_data.groupby("Gender").sum()["Price"]
gender_purchase_val

#avg purchase per person
gender_average_purchase = gender_purchase_val / total_gender
gender_average_purchase

#Gender purchasing analysis Dataframe
gender_purchasing_demographics = pd.DataFrame({"Purchase Count": gender_purchases,
                                "Average Purchase Price": gender_averages.astype(float).map("${:,.2f}".format),
                                "Total Purchase Value": gender_purchase_val.astype(float).map("${:,.2f}".format),
                                "Avg Total Purchase per Person": gender_average_purchase.astype(float).map("${:,.2f}".format)})
gender_purchasing_demographics
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Find min/max ages as a check for low and high ends of age range
# print(purchase_data["Age"].max())
# print(purchase_data["Age"].min())

#Establish Age Bins
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]
age_bins

#Bin lables
age_groups = ["<10", "10-14", "15-19", "20-24","25-29","30-34","35-39","40+"]
age_groups

#Slice the data into the bins
pd.cut(purchase_data["Age"], age_bins, labels=age_groups).head()

#Place age bins into the DataSeries
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], age_bins, labels=age_groups)
purchase_data.head()

#Create a GroupBy object for "Age Group"
age_group = purchase_data.groupby("Age Group")
age_group

#Customer Demographics
age_group_demo = purchase_data.loc[:,["SN","Age Group"]]
age_group_demo

#Remove SN duplicates
age_group_demo_nodups = age_group_demo.drop_duplicates()
age_group_demo_nodups

#Count of players by Age Group
total_age_group = age_group_demo_nodups["Age Group"].value_counts()
total_age_group

#Percentage of players by Age Group
age_group_percentage = total_age_group / total_unique_players * 100
age_group_percentage

#Summarize DataFrame
age_demographics = pd.DataFrame({"Total Count": total_age_group,
                                "Percentage of Players": age_group_percentage.astype(float).map("{:,.2f}%".format)})
age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Purchases by Age Group
age_group_purchases = purchase_data["Age Group"].value_counts()
age_group_purchases

#Average purchase price by Age Group
age_group_averages = purchase_data.groupby("Age Group").mean()["Price"]
age_group_averages

#Total purchase value by Age Group
age_group_purchase_val = purchase_data.groupby("Age Group").sum()["Price"]
age_group_purchase_val

#avg purchase per person
age_group_average_purchase = age_group_purchase_val / total_age_group
age_group_average_purchase

#Summarize Age Group Purchase Data
age_purchasing_analysis = pd.DataFrame({"Purchase Count": age_group_purchases,
                                       "Average Purchase Price": age_group_averages.astype(float).map("${:,.2f}".format),
                                       "Total Purchase Value": age_group_purchase_val.astype(float).map("${:,.2f}".format),
                                       "Average Price per Person": age_group_average_purchase.astype(float).map("${:,.2f}".format)})
age_purchasing_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Count up purchases by player
player_purchases = purchase_data["SN"].value_counts()
player_purchases

#Avg price per person
avg_player_purchase = purchase_data.groupby("SN").mean()["Price"]
avg_player_purchase

#Total purchase value by person
player_purchase_val = purchase_data.groupby("SN").sum()["Price"]
player_purchase_val

#Player Spending Summary Dataframe
player_spending_summary = pd.DataFrame({"Purchase Count": player_purchases,
                            "Average Purchase Price": avg_player_purchase.astype(float).map("${:,.2f}".format),
                            "Total Purchase Value": player_purchase_val})
player_spending_summary

#Sort DataFrame by top spender Total Purchase Value
top_spenders_sorted = player_spending_summary.sort_values(["Total Purchase Value"], ascending=False)
top_spenders_sorted.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Item Count
item_count = purchase_data.groupby(['Item ID','Item Name']).count()["Price"]
item_count

#Item Prices
item_prices = purchase_data.groupby(['Item ID','Item Name']).mean()["Price"]
item_prices

#Item Purchases
item_purchases = purchase_data.groupby(['Item ID','Item Name']).sum()["Price"]
item_purchases

#Item Summmary DataFrame
item_summary = pd.DataFrame({"Purchase Count": item_count,
                             "Item Price": item_prices.astype(float).map("${:,.2f}".format),
                            "Total Purchase Value": item_purchases.astype(float).map("${:,.2f}".format)})
item_summary.head()

#Sort by item Purchase Count
purchase_sort = item_summary.sort_values(["Purchase Count"], ascending=False)
purchase_sort.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Establish most profitable items DataFrame
most_profiitable_items = pd.DataFrame({"Purchase Count": item_count,
                             "Item Price": item_prices.astype(float).map("${:,.2f}".format),
                            "Total Purchase Value": item_purchases})
most_profiitable_items

#Sort by most profitable items Total Purchase Value
profitable_item_sort = most_profiitable_items.sort_values(["Total Purchase Value"], ascending=False)
profitable_item_sort.head()
```
